1. Create a single dataframe with the concatenation of all input csv files, adding a column called `country`.

```python
import pandas as pd

ca = pd.read_csv(r'Files\CAvideos.csv', parse_dates=['publish_time'])
de = pd.read_csv(r'Files\DEvideos.csv', parse_dates=['publish_time'])
fr = pd.read_csv(r'Files\FRvideos.csv', parse_dates=['publish_time'])
gb = pd.read_csv(r'Files\GBvideos.csv', parse_dates=['publish_time'])
ind = pd.read_csv(r'Files\INvideos.csv', parse_dates=['publish_time'])
jp = pd.read_csv(r'Files\JPvideos.csv', parse_dates=['publish_time'], encoding='iso-8859-1')
kr = pd.read_csv(r'Files\KRvideos.csv', parse_dates=['publish_time'], encoding='iso-8859-1')
mx = pd.read_csv(r'Files\MXvideos.csv', parse_dates=['publish_time'], encoding='iso-8859-1')
ru = pd.read_csv(r'Files\RUvideos.csv', parse_dates=['publish_time'], encoding='iso-8859-1')
us = pd.read_csv(r'Files\USvideos.csv', parse_dates=['publish_time'])
```

```python
ca['country'] = 'CA'
de['country'] = 'DE'
fr['country'] = 'FR'
gb['country'] = 'GB'
ind['country'] = 'IND'
jp['country'] = 'JP'
kr['country'] = 'KR'
mx['country'] = 'MX'
ru['country'] = 'RU'
us['country'] = 'US'
```

```python
countries = [ca, de, fr, gb, ind, jp, kr, mx, ru, us]
```

```python
df = pd.concat(countries, ignore_index=True)
df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')
df
```

2. Extract all videos that have no tag.

```python
df[df['tags'] == '[none]']
```

3. For each channel, determine the total number of views.

```python
total_views = df.groupby('channel_title')['views'].sum()
total_views
```

4. Save all rows with disabled comments and disabled ratings, or that have `video_error_or_removed` in a new dataframe called `excluded`, and remove those rows from the original dataframe.

```python
excluded = df[(df['comments_disabled'] == True) & (df['ratings_disabled'] == True) | (df['video_error_or_removed'] == True)]
df = df.drop(excluded.index)
df
```

5. Add a `like_ratio` column storing the ratio between the number of likes and of dislikes

```python
like_ratio = df['likes']/ (df['likes'] + df['dislikes'])
df['like_ratio'] = like_ratio
df
```

6. Cluster the publish time into 10-minute intervals (e.g. from 02:20 to 02:30)

```python
df['publish_interval'] = df['publish_time'].dt.floor('10min')
df
```

7. For each interval, determine the number of videos, average number of likes and of dislikes.

```python

task_7 = df.groupby('publish_interval').agg(
    {'video_id':'count',
    'likes':'mean',
    'dislikes':'mean'}
)
task_7
```

8. For each tag, determine the number of videos.
Notice that `tags` contains a string with several tags.

```python
tags_exploded = df['tags'].str.split('|').explode()
tags_count = tags_exploded.value_counts()
tags_count
```

9. Find the tags with the largest number of videos

```python
tags_count.drop('[none]').head(10)
```

10. For each (`tag`, `country`) pair, compute average ratio likes/dislikes

```python
temporary_df = df[['tags', 'country', 'like_ratio']].copy()
temporary_df['tags'] = temporary_df['tags'].str.split('|')
temporary_df = temporary_df.explode('tags')
ratio = temporary_df.groupby(['tags','country'])['like_ratio'].mean()
ratio = ratio.reset_index()
ratio
```

11. For each (`trending_date`, `country`) pair, the video with the largest number of views

```python
winner_id = df.groupby(['trending_date', 'country'])['views'].idxmax()
most_watched = df.loc[winner_id]
most_watched
```
